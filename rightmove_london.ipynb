```python
import requests
import datetime
from rightmove_webscraper import RightmoveData
import numpy as np
```

```python
url_10_miles = "https://www.rightmove.co.uk/property-to-rent/find.html?locationIdentifier=REGION%5E61224&maxBedrooms=2&minBedrooms=2&maxPrice=1500&minPrice=1000&radius=10.0&propertyTypes=&includeLetAgreed=false&mustHave=&dontShow=houseShare&furnishTypes=partFurnished%2Cfurnished&letType=longTerm&keywords="
```

```python
rightmove_10miles = RightmoveData(url_10_miles)
```

```python
rightmove_10miles.summary()
```

```python
flatsWithin10Miles = rightmove_10miles.get_results
```

```python
def google_maps_query(origin, destination, mode, api_key):
    url = "https://maps.googleapis.com/maps/api/distancematrix/json?"

    params = {"key" : api_key, 
             "origins" : origin,
              "destinations" : destination,
             "departure_time" : int(datetime.datetime(2021, 9, 2, 9, 0, 0, 0).timestamp()),
             "mode" : mode,
             "units" : "metric"}
    payload={}
    headers = {}
    response = requests.request("GET", url, headers=headers, data=payload, params=params)
    return response
    
    
def get_commute_times(origin, destination, api_key):
    times = {"distance" : np.nan, "transit" : np.nan, "bicycling" : np.nan, "walking" : np.nan}
    
    for mode in ["transit", "bicycling", "walking"]:
        response = google_maps_query(origin, destination, mode)
        if response.status_code == 200:
            response_json = response.json()
            if len(response_json["rows"]) != 1:
                print("wrong number of rows!")
            try:
                times[mode] = response_json["rows"][0]["elements"][0]["duration"]["value"]
                times["distance"] = response_json["rows"][0]["elements"][0]["distance"]["value"]
            except:
                print("Error!")
                
    return times
           
get_commute_times("Hill House, Defence Close, SE28", "UCL Main Campus, London")
# resp = google_maps_query("Hill House, Defence Close, SE28", "UCL Main Campus, London", "walking") 
```

```python
transit = np.empty((len(flatsWithin10Miles),))
transit[:] = np.nan
bicycling = np.empty((len(flatsWithin10Miles),))
bicycling[:] = np.nan
walking = np.empty((len(flatsWithin10Miles),))
walking[:] = np.nan
distance = np.empty((len(flatsWithin10Miles),))
distance[:] = np.nan
```

```python
for idx, row in flatsWithin10Miles.iterrows():
    origin = row["address"]
    
    commute_times = get_commute_times(origin, "UCL Main Campus, London")
    transit[idx] = commute_times["transit"]
    bicycling[idx] = commute_times["bicycling"]
    walking[idx] = commute_times["walking"]
    distance[idx] = commute_times["distance"]
```
