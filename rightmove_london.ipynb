```python
def google_maps_query(origin, destination, mode, api_key):
    url = "https://maps.googleapis.com/maps/api/distancematrix/json?"

    params = {"key" : api_key, 
             "origins" : origin,
              "destinations" : destination,
             "departure_time" : int(datetime.datetime(2021, 9, 9, 9, 0, 0, 0).timestamp()),
             "mode" : mode,
             "units" : "metric"}
    payload={}
    headers = {}
    response = requests.request("GET", url, headers=headers, data=payload, params=params)
    return response
    
    
def get_commute_times(origin, destination, api_key):
    times = {"distance" : np.nan, "transit" : np.nan, "bicycling" : np.nan, "walking" : np.nan}
    
    for mode in ["transit", "bicycling", "walking"]:
        response = google_maps_query(origin, destination, mode, api_key)
        if response.status_code == 200:
            response_json = response.json()
            if len(response_json["rows"]) != 1:
                print("wrong number of rows!")
                print(response_json)
            try:
                times[mode] = response_json["rows"][0]["elements"][0]["duration"]["value"]
                times["distance"] = response_json["rows"][0]["elements"][0]["distance"]["value"]
            except:
                print("Error!")
                
    return times
           
get_commute_times("Grampian Gardens, Golders Green Estate", "UCL Main Campus, London", finder.api_key)
# resp = google_maps_query("Hill House, Defence Close, SE28", "UCL Main Campus, London", "walking") 
```
